# Observations data for training

Climetlib provides the observation datasets. They can be used as a xarray.Dataset :

```python
cmlds = cml.load_dataset("s2s-ai-challenge-observations",
                         freq='daily',
                         parameter='t2m'                         
                       )
cmlds.to_xarray()
```

```python
cmlds = cml.load_dataset("s2s-ai-challenge-observations",
                         freq='weekly',
                         parameter='t2m'                         
                       )
cmlds.to_xarray()
```
